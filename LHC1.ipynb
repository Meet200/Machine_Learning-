Analyse the growth rate of covid-19 in INDIA with referenceto number of cases, deaths and recovered. You may use the 7-days moving average for better graph visualization

```python
# confirmed cases
dataset = pd.read_csv("time_series_covid19_confirmed_global.csv")
india= dataset[dataset['Country/Region']=='India']
indi= india.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
indi.index = ["Confirmed"]
d = indi.transpose()

# Recovered cases
dataset2 = pd.read_csv("time_series_covid19_recovered_global.csv")
india2= dataset2[dataset2['Country/Region']=='India']
indi2= india2.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
indi2.index = ["Recover"]
d2 = indi2.transpose()
d['Recover'] = d2

# Death cases
dataset3 = pd.read_csv("time_series_covid19_deaths_global.csv")
india3= dataset3[dataset3['Country/Region']=='India']
indi3= india3.drop(['Province/State', 'Country/Region','Lat','Long'], axis = 1)
indi3.index = ["Death"]
d3= indi3.transpose()
d['Death'] = d3
d
```

```python
plt.plot(d.index, d['Confirmed'], label = "Confirmed", linestyle="-")
plt.plot(d.index, d['Death'], label = "Deaths", linestyle="--")
plt.plot(d.index, d['Recover'], label = "Recovered", linestyle="-.")
plt.xlabel("Days")
plt.ylabel("Number of Cases")
plt.title("Indian Covid Cases")
plt.legend()
plt.show()
```

Task-4 Analyse the cases of Confirmed, Death and Recovered figures from 01 March, 2021 in Gujarat

```python

import os
import glob
import pandas as pd

location = "LHC"

file_name = []
for file in os.listdir(location):
         file_name.append(file)
         
print(file_name)

#combine all files in the list
combined_csv = pd.concat([pd.read_csv("LHC/"+f) for f in file_name ])
#export to csv
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
data_india = combined_csv.loc[combined_csv['Country_Region']=='India']

data_india

data_india_gujrat = data_india.loc[data_india['Province_State']=='Gujarat']

data_india_gujrat.head(10)
data_india_gujrat = data_india_gujrat.sort_values(by=['Last_Update'])
data_india_gujrat

plt.plot(data_india_gujrat['Last_Update'], data_india_gujrat['Deaths'], label = "Deaths")
plt.plot(data_india_gujrat['Last_Update'], data_india_gujrat['Recovered'], label = "recovered")
plt.plot(data_india_gujrat['Last_Update'], data_india_gujrat['Confirmed'], label = "confirmed")
plt.xlabel("Days ")
plt.ylabel("cumulative Cases Death,Recovered,Confirmed")
plt.title("Gujarat Coivd status after  1st March 2021")
plt.legend()
plt.show()
```

```python
# Using Linear regression
data_pred_guj = [i for i in range(1,65)]
df2 = data_india_gujrat
df2.insert(0,"Days",data_pred_guj,True)
df2

X_features = df2.iloc[:,0:1]
Y_features = df2.iloc[:,8:9]

print(Y_features)

from sklearn.linear_model import LinearRegression
prediction_model = LinearRegression().fit(X_features, Y_features)
x = np.array([65]).reshape(-1, 1)
x

y_pred_gujrat = prediction_model.predict(x)
print(int(y_pred_gujrat))

```

```python
from sklearn.linear_model import LinearRegression
prediction_model = LinearRegression().fit(X_features, Y_features)
x = np.array([65]).reshape(-1, 1)
x

y_pred_gujrat = prediction_model.predict(x)
print(int(y_pred_gujrat))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X_features)
poly = poly.fit(X_poly, Y_features)
lin2 = LinearRegression()
lin2.fit(X_poly, Y_features)
x = np.array([65]).reshape(-1, 1)
x2 = np.array([64]).reshape(-1, 1)
ans2 =int(lin2.predict(poly.fit_transform(x))) 
ans3 =int(lin2.predict(poly.fit_transform(x2))) 
print(ans2)
print(ans3)
```

```python
print("Total Case in gujrat on 4th May : ",(ans2-ans3))
fin_guj = ans2-ans3
```

```python
perct = float((fin_guj/(fin_ind))*100)
print(perct)
plt.pie([perct,100-perct],labels=['Gujarat','India - Gujrat'],colors=["orange","red"],
        explode=(0.1,0),shadow=True,autopct='%1.1f%%')
plt.title(" 4th May case comparison")
plt.legend()
plt.show()
print("number of cases in gujarat",13239)
print("number of cases in india",359940)
```
